Now going to implement the second part of this, which consists of a smoothing filter to transform the non-periodic square wave of the histogram to the non-periodic sine wave $H_s(y)$ using the following method:
$H_s(y) = {1\over m}\left(\sum^{y+\lceil m/2 \rceil-1}_{k=y-\lfloor m/2 \rfloor -1}H(k)\right)$

$m$ here is the size of the smoothing window, which is determined through several steps - first I need to generate a periodic square wave with period $\tau$:

$R_\tau(y) = max[H(y)]\cdot sgn\left[sin({{2\pi y}\over{\tau}})+1\right]$

```python
def periodic_square_wave(half_amp, period, position):
    return half_amp*np.sign(np.sin((2*np.pi*position)/period)+1)

def gen_sqr_wave(arr, period):
    """ 1. Generate a square wave with the given period from an array. """
    amp = float(arr.max())
    return np.asarray([periodic_square_wave(amp, period, a)for a in arr])

def variance(hist, period):
    """ 2. Calculate the variance between the histogram and a square wave generated from it. """
    return (1.0/horizontal_hist.size)*np.sum((gen_sqr_wave(horizontal_hist, period) - horizontal_hist)**2)

def min_variance(hist):
    all_vars = np.asarray([variance(hist, i) for i in range(1,hist.size/2)])
    return np.argmax(all_vars)+1

def window_size(hist):
    """ 3.  """
    return min_variance(hist)

def smoothing_filter(hist):
    window = window_size(hist)
    smoothed = np.empty_like(hist)
    for i in range(hist.size):
        from_ind = i-np.floor(window/2)
        to_ind = i+np.ceil(window/2)-1
        sect = hist[from_ind:to_ind]
        smoothed[i] = (1.0/window) * np.sum(sect)
    return smoothed
        
```

```python
smooth_hist = smoothing_filter(horizontal_hist)
smooth_display_hist = smooth_hist - smooth_hist.min()
```

With that smoothing filter written, I now need to detect all inflection points in the histogram - these will apparently correspond with the boundaries between characters and background regions. I think that it probably won't do too bad a job with this example.

```python
deriv = np.empty_like(smooth_hist)
for i in range(smooth_hist.size-1):
    deriv[i] = (smooth_hist[i+1] - smooth_hist[i])
```

```python
np.argmin(deriv)
```

```python
seq_Y = []
for i in range(deriv.size):
    if deriv[i] < 50 and deriv[i] > -50:
        seq_Y.append(i)
```
